This example will demonstrate using [GeoPandas](https://github.com/kjordahl/geopandas) to load and plot some polygons representing the New York City boroughs. Then we'll put it into a Leaflet map right in the notebook using [mplleaflet](https://github.com/jwass/mplleaflet).

```python
%matplotlib inline
import os

import geopandas as gpd
import matplotlib.pyplot as plt

import mplleaflet
```

```python
filename = os.path.join('data', 'nybb.zip')
df = gpd.read_file('/nybb.shp', vfs='zip://' + filename)
```

```python
ax = df.plot()
```

```python
# The display call inserts the html in the IPython notebook.
# An interactive, panable, zoomable map.
mplleaflet.display(fig=ax.figure, crs=df.crs)
```
